```python
import numpy as np
import pylab as pl


data=np.genfromtxt('Spike_times.txt')


select = np.array([d for d in data if d[1] < 30])

data1= select.transpose()

x = [0.1*d for d in data1[0]]
y = [d for d in data1[1]]

print (data1[0].size)
print ("with 0.5")

pl.scatter(x, y, s=25,edgecolors='none')

pl.show()
```
